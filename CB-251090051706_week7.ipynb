```python
# (1)Matplot library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("mtcars.csv")
# Display all rows and columns
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',None)
```

```python
df
```

```python
mpg=df['mpg']
mpg.plot()
```

```python
# (2)Plot any column about an index
df=df[['cyl','wt','mpg']]
df.plot()
```

```python
# (3)Bar graph
cyl_counts=df['cyl'].value_counts()
cyl_counts.plot(kind='bar',color='green')
plt.title("Count of Cars by Cylinders.")
plt.xlabel("Cylinders")
plt.ylabel("Count")
plt.show()
```

```python
# (4)Stylizing the graphs
plt.style.use('ggplot')
```

```python
# (4)Scatter plot
df.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black');
```

```python
# (5)Box plot
plt.figure(figsize=(6,6))
sns.boxplot(x='cyl',y='mpg',data=df)
plt.title("MPG by Cylinder Category.")
plt.xlabel("Number of Cylinders")
plt.ylabel("MPG")
plt.show()
```

```python
# (6)Histogram
x=df['mpg']
plt.hist(x)
plt.title("Distribution of MPG in Cars Dataset.")
plt.xlabel("MPG")
plt.ylabel("Frequency")
```

```python
# (7)Pie Chart
cyl_counts=df['cyl'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(
     cyl_counts,
     labels=cyl_counts.index,
     autopct='%1.1f%%',#Add percentage labels
    colors=['skyblue','lightgreen','orange'],
    startangle=140,# Starts angle of pie chart
    explode=[0.1,0,0]
)
plt.title('Proportion of Cars by Number of Cylinders')
plt.show()
```

```python
df=pd.read_csv('mtcars.csv')
am_counts=df['am'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(
    am_counts,
    labels=['Automatic','Manual'],
    autopct="%1.1f%%",
    colors=['red','blue'],
    startangle=90
)
plt.title("Proportion of Cars by Transmission Tyre.")
plt.show()
```

```python
# (8)Nested Pie
outer_labels=df['cyl'].value_counts().index
outer_sizes=df['cyl'].value_counts().values
inner_labels=df['gear'].value_counts().index
inner_sizes=df['gear'].value_counts().values
fig,ax=plt.subplots(figsize=(8,8))
ax.pie(outer_sizes,labels=outer_labels,radius=1,autopct="%1.1f%%",colors=['purple','lightblue','orange'])
ax.pie(inner_sizes,labels=inner_labels,radius=0.7,autopct="%1.1f%%",colors=['green','gold','lightcoral'])
ax.set(aspect="equal")
plt.title("Nested Pie Chart:Cylinders(outside) and Gears(inside)")
plt.show()
```

```python
# Q.1 Load the IPL dataset and display first 10 records
# Q.2 Find total no. of matches,teams and seasons
# Q.3 Plot no. of matches played per season
# Q.4 Use seaborn to acess top 5 teams with most wins
# Q.5 Create a box plot showing top 10 run scorrer
# Q.6 Plot a pie chart showing match wins per venue
# Q.7 Compare home v/s away wins by suitable plot
# Q.8 Create a visualization whic tells a IPL performance story
# Q.9. Add headlines and stories
```

```python
# Q.1 answer
ipl=pd.read_csv("data.csv")
print(ipl[:10])
```

```python
# Q.2 answer
print("Total no. of matches:",ipl.shape[0])
print("Total unique teams:",len(pd.unique(ipl['team1'].append(ipl['team2']))))
print("Total seasons:",ipl["season"].nunique())
```

```python
# Q.3 answer
x=ipl["season"]
plt.hist(x,bins=ipl["season"].nunique())
plt.title("Distribution of Matches per Seasons.")
plt.xlabel("Seasons")
plt.ylabel("Matches")
```

```python
# Q.4 answer
top_wins = ipl['winner'].value_counts().head(5)
plt.figure(figsize=(12,12))
sns.barplot(x=top_wins.index, y=top_wins.values)
plt.title("Top 5 teams with most wins.")
plt.ylabel("Wins")
plt.show()
```

```python
# Q.5 answer
top_run_matches = ipl.nlargest(10, 'win_by_runs')
plt.figure(figsize=(8,5))
sns.boxplot(x=top_run_matches['winner'], y=top_run_matches['win_by_runs'])
plt.title("Top 10 Run Scorers")
plt.xticks(rotation=45)
plt.show()
```

```python
# Q.6 answer
venue_wins = ipl['venue'].value_counts()
plt.figure(figsize=(40,40))
plt.pie(venue_wins.values, labels=venue_wins.index, autopct='%1.1f%%')
plt.title("Match Wins per Venue")
plt.show()
```

```python
# Q.7 answer
home_wins = (ipl['winner'] == ipl['team1']).sum()
away_wins = (ipl['winner'] == ipl['team2']).sum()
plt.bar(['Home Wins','Away Wins'], [home_wins, away_wins], color=['green','orange'])
plt.title("Home vs Away Wins")
plt.show()
```

```python
# Q.8 answer
batsmen = [
    "Brendon McCullum", "Sourav Ganguly", "Ricky Ponting",
    "David Hussey", "Mohammad Hafeez"
]
runs = [158, 10, 20, 12, 5]
plt.figure(figsize=(20,20))
plt.bar(batsmen, runs, color="purple", edgecolor="black")
plt.title("Runs Scored by Each Batsman - IPL Opening Match (KKR vs RCB, 2008)")
plt.xlabel("Batsmen")
plt.ylabel("Runs Scored")
plt.show()
```
